```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplot(df, feats):
    for feature in feats:
        tdf = df[[feature, "label"]].compute()
        print("Boxplot for", feature)
        plt.figure(figsize=(10,6))
        sns.boxplot(x='label', y=feature, data=tdf)
        plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def logscale_boxplot(df, feats):
    for feature in feats:
        tdf = df[[feature, "label"]].compute()
        print("Boxplot for", feature)
        plt.figure(figsize=(10,6))
        ax = sns.boxplot(x='label', y=feature, data=tdf)
        if "corr" in feature:
            ax.set(yscale="symlog")
        else:
            ax.set(yscale="log")
        plt.show()
```

```python
logscale_boxplot(ddf, features_10)
```
