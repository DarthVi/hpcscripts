```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplot(df, feats):
    for feature in feats:
        tdf = df[[feature, "label"]]
        print("Boxplot for", feature)
        plt.figure(figsize=(10,6))
        sns.boxplot(x='label', y=feature, data=tdf)
        plt.show()
```

```python
def logscale_boxplot(df, feats):
    for feature in feats:
        tdf = df[[feature, "label"]]
        print("Boxplot for", feature)
        plt.figure(figsize=(10,6))
        ax = sns.boxplot(x='label', y=feature, data=tdf)
        if "corr" in feature:
            ax.set(yscale="symlog")
        else:
            ax.set(yscale="log")
        plt.show()
```

```python
some_features_RFE = ["perc75_ctxt", "max_opaif0/portXmitPackets", "perc5_max_cpus/cache-references",
           "perc25_thp_fault_alloc", "perc5_max_cpus/branch-instructions", "mean_max_cpus/cache-references",
           "std_AnonPages", "perc95_Active", "median_max_cpus/branch-instructions", "std_Active-anon",
                    "perc95_max_cpus/cache-references"]
```

```python
boxplot(df, some_features_RFE)
```
